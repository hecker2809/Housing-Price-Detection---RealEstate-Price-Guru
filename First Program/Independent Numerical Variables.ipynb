```python
import pandas as pd
raw_data = pd.read_csv('Raw_Housing_Prices2.csv')
raw_data
```

```python
numerical_column = ['No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'Area of the House from Basement (in Sqft)',
                   'Latitude', 'Longitude', 'Living Area after Renovation (in Sqft)']
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'median')

raw_data[numerical_column] = imputer.fit_transform(raw_data[numerical_column])
```

```python
raw_data.info()
```

```python
# For ZIPCODE as it should be treated as categorical type

imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
raw_data['Zipcode'] = imputer.fit_transform(raw_data['Zipcode'])
```

```python
# To overcome this error

column = raw_data['Zipcode'].values.reshape(-1,1)
column.shape

imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
raw_data['Zipcode'] = imputer.fit_transform(column)
```

```python
raw_data1['Zipcode'] = raw_data['Zipcode'].astype(object)
raw_data1.dtypes
```

```python
# Assuming you have a DataFrame 'raw_data'

# Replace NaN values in the 'No of Times Visited' column with 'None'
raw_data1['No of Times Visited'].fillna('None', inplace=True)

# Now, 'None' will replace all NaN values in the column
print(raw_data1['No of Times Visited'].unique())


mapping = {'None':'0','Once':'1','Twice':'2','Once':'1','Thrice':'3','Four':'4' }

raw_data1['No of Times Visited'] = raw_data1['No of Times Visited'].map(mapping)

print(raw_data1['No of Times Visited'].unique())
```

```python
raw_data1['Ever Renovated']  = np.where(raw_data1['Renovated Year'] == 0, 'No','Yes')
raw_data1.head()
```

```python
raw_data1['Purchase_Year'] = pd.DatetimeIndex(raw_data1['Date House was Sold']).year

raw_data1['Years Since Renovation'] = np.where(raw_data1['Ever Renovated'] == 'Yes',
                                              abs(raw_data1['Purchase_Year']-raw_data1['Renovated Year']),0)
raw_data1.head()
```

```python
raw_data1.drop(columns = ['Purchase_Year','Date House was Sold','Renovated Year'],inplace = True)
raw_data1.head()
```
